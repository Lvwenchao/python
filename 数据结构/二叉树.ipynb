### 1. 二叉树的前中后遍历
- 前序遍历：中左右
- 中序遍历：左中右
- 后序遍历：左右中
- 层序遍历: 每个节点的左右子树

```python
class BTree(object):
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    def preorder(self):
        result = ""
        if self.val is not None:
            result += self.val
        if self.left is not None:
            result += self.left.preorder()
        if self.right is not None:
            result += self.right.preorder()
        return result

    def inorder(self):
        if self.left is not None:
            self.left.inorder()
        if self.val is not None:
            print(self.val, end=' ')
        if self.right is not None:
            self.right.inorder()

    def postorder(self):
        if self.left is not None:
            self.left.inorder()
        if self.right is not None:
            self.right.inorder()
        if self.val is not None:
            print(self.val, end=' ')

    # 层序遍历
    def levelorder(self, root):
        # 层序遍历列表
        re = []
        nodes = []
        if root:
            nodes = [root]
        while nodes:
            temp_vals = []
            temp_nodes = []
            for node in nodes:
                # 记录中节点
                if node.val is not None:
                    temp_vals.append(node.val)
                # 左节点存在
                if node.left is not None:
                    temp_nodes.append(node.left)
                # 右节点存在
                if node.right is not None:
                    temp_nodes.append(node.right)
            nodes = temp_nodes
            re.append(temp_vals)
        return re

    def height(self):

        if self.val is None:
            return 0
        elif self.left is None and self.right is None:
            return 1
        elif self.left is not None and self.right is None:
            return 1 + self.left.height()
        elif self.left is None and self.right is not None:
            return 1 + self.right.height()
        else:
            return 1 + max(self.left.he(), self.right.height())


leftree = BTree('f')
leftree.left = BTree('h')
leftree.right = BTree('i')

right_tree = BTree('g')
right_tree.left = BTree('j')
right_tree.right = BTree('k')

Tree = BTree('c')
Tree.left = leftree
Tree.right = right_tree

right_tree = Tree
leftree = BTree('b')
leftree.left = BTree('d')
leftree.right = BTree('e')
Tree = BTree('a')

Tree.left = leftree
Tree.right = right_tree

result = Tree.preorder()
print(result)
```
